```python
#Make a Code Cell that imports numpy as np, matplotlib.pyplot as plt, and scipy.stats as stats.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Make a Code Cell that imports your v_rest_sst.csv and v_rest_pvalb.csv variables as numpy arrays called v_rest_sst and v_rest_pvalb respectively

url1='https://raw.githubusercontent.com/mpompy/Neur265/main/v_rest_sst.csv'
v_rest_sst=np.loadtxt(url1)

url2='https://raw.githubusercontent.com/mpompy/Neur265/main/v_rest_pvalb.csv'
v_rest_pvalb=np.loadtxt(url2)
```

```python
v_rest_sst_mean=np.mean(v_rest_sst)
print(v_rest_sst_mean)
v_rest_pvalb_mean=np.mean(v_rest_pvalb)
print(v_rest_pvalb_mean)


t_test, p_value=stats.ttest_ind(v_rest_sst_mean, v_rest_pvalb_mean)


if p_value>0.05:

  fig, ax = plt.subplots(1,1)
  plt.xlabel('resting membrane potential')
  plt.ylabel('frequency')
  ax[0].hist(v_rest_sst)
  ax[1].hist(v_rest_pvalb)

else:
   print('no graph')

```

```python
# Code Cell that defines a function called vrestHist(). This function should execute the code in step 4 when called.

def vrestHist():

  if p_value>0.05:
    fig, ax = plt.subplots(1,1)
    plt.xlabel('resting membrane potential')
    plt.ylabel('frequency')
    ax[0].hist(v_rest_sst)
    ax[1].hist(v_rest_pvalb)

  else:
   print('no graph')

```

```python
#Make a Code Cell that calls vrestHist().

vrestHist()
```
