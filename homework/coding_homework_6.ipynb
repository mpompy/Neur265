```python
#Code Cell that imports relevant modules - specifically, numpy as np, and matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Code Cell that imports these files as numpy arrays called acc_hits, acc_fa, lc_hits, and lc_fa, respectively.

url1='https://raw.githubusercontent.com/mpompy/Neur265/main/acc_hits_hw6.csv'
url2='https://raw.githubusercontent.com/mpompy/Neur265/main/acc_fa_hw6.csv'
url3='https://raw.githubusercontent.com/mpompy/Neur265/main/lc_hits_hw6.csv'
url4= 'https://raw.githubusercontent.com/mpompy/Neur265/main/lc_fa_hw6.csv'

acc_hits=np.loadtxt(url1,delimiter = ',')
acc_fa=np.loadtxt(url2,delimiter = ',')
lc_hits=np.loadtxt(url3,delimiter = ',')
lc_fa=np.loadtxt(url4,delimiter = ',')


```

```python
#Code Cell that prints the shape of each array

print(acc_hits.shape)
print(acc_fa.shape)
print(lc_hits.shape)
print(lc_fa.shape)
```

```python
#Code Cell,that  creates a variable called time that has the same length as your four arrays, with equally spaced values between "zero" and "four" (the length of each trial, in seconds).

time=np.linspace(0,4,8001)
```

```python
#Code Cell, that creates an expression to determine the sampling rate of the data

number_of_samples=8001
total_duration=4*8001

sampling_rate=number_of_samples/total_duration

print(sampling_rate)
```

```python
#Code Cell, with plot of the first trial of acc_hits in red, and the first trial of lc_hits in the same plot in blue in one subplot. In a second subplot, plot the first trial of acc_fa in red, and the first trial of lc_fa in blue.

fig, axs=plt.subplots(2,1,figsize=(10,10))

axs[0].plot(time,acc_hits[0,:],color='red')
axs[0].plot(time,lc_hits[0,:],color='blue')
axs[0].set_xlabel('time')
axs[0].set_ylabel('voltage [$\mu V$]')
plt.legend(['acc_hits', 'lc_hits'])
plt.title('acc_hits & lc_hits')



axs[1].plot(time,acc_fa[0,:],color='red')
axs[1].plot(time,lc_fa[0,:],color='blue')
axs[1].set_xlabel('time')
axs[1].set_ylabel('voltage [$\mu V$]')
plt.title('acc_fa & lc_fa')
plt.legend(['acc_fa', 'lc_fa'])
```

```python
#Code Cell, that computes the coherence for acc_hits and lc_hits

dt = time[1] - time[0]
N = acc_hits.shape[1]

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])


T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)


cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real,'r')
plt.xlim([0, 50])                         ,
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
```

 In which frequency band is there the most phase coherence during hits

 - There is the most phase coherence during hits around 2 Hz. At this band we see a sharp peak.

```python
#Code Cell, that computes the coherence for acc_fa and lc_fa

dt = time[1] - time[0]
N = acc_fa.shape[1]

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])


T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)


cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real,'b')
plt.xlim([0, 50])                         ,
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
```

Interpretation of my plot from step 12.
