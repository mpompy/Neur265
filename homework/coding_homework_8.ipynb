```python
#Code Cell that imports relevant modules - specifically, numpy as np, matplotlib.pyplot as plt, and from scipy.stats import pearsonr

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
#Code cell that imports all of these files to your GitHub repo, and import them as numpy arrays named fear_traces, extinction_predictors, shock_index, and tone_index, respectively, in a Code Cell.

fear_traces=np.loadtxt('https://raw.githubusercontent.com/mpompy/Neur265/main/fear_conditioning_traces_hw8.csv', delimiter = ',')
extinction_predictors=np.loadtxt('https://raw.githubusercontent.com/mpompy/Neur265/main/extinction_predictors_hw8.csv', delimiter = ',')
shock_index=np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int)
tone_index=np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int)
```

```python
#Code Cell,that makes two new variables - conditioning_traces, and extinction_traces

conditioning_traces=fear_traces[0:10458,:]
extinction_traces=fear_traces[10459:,:]
```

```python
#Code Cell, that prints the size of your conditioning_traces and extinction_traces arrays.

print(conditioning_traces.size)
print(conditioning_traces.shape)
print(extinction_traces.size)
print(extinction_traces.shape)
```

```python
# Code Cell, that makes a predictors_conditioning variable that contains shock onset and tone onset.


shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0


predictors_conditioning=df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array




```

```python
#. In a new Code Cell, create your r_model_conditioning variable, and run your for loop.

r_model_conditioning= np.zeros(63)
result_list=[]

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r3.statistic
  result_list.append(r_model_conditioning)

r_index_conditioning = np.array(result_list)




```

```python
#new Code Cell that run another loop that does the same thing for your extinction_traces array as the conditioning_traces array


shock_array2 = np.zeros(14209)
for i in range(14209):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array2[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array2[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array2[i] = 1
  else:
    shock_array2[i] = 0

tone_array2 = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array2[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array2[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array2[i] = 1
  else:
    tone_array2[i] = 0


predictors_extinction=df(data={'Intercept': np.ones_like(tone_array2), 'tone onset': tone_array2})
predictors_extinction['shock onset'] = shock_array2

r_model_extinction= np.zeros(63)
result_list2=[]

for i in range(63):
  trace_temp2 = extinction_traces[:,i]
  model31 = sm.GLM(trace_temp2, predictors_extinction, family=Poisson())
  model31_results = model31.fit()
  predicted_trace1 = model31_results.predict()
  r31 = pearsonr(trace_temp2, predicted_trace1)
  r_model_extinction[i] = r31.statistic
  result_list2.append(r_model_extinction)

r_index_extinction = np.array(result_list2)
```

```python
# Code Cell,that uses the pearsonr function to correlate your r_index_conditioning and r_index_extinction variables


r_index_conditioning1 = r_index_conditioning[~np.isnan(r_index_conditioning)]
r_index_extinction1 = r_index_extinction[~np.isnan(r_index_extinction)]
correlation_coefficient = pearsonr(r_index_conditioning1, r_index_extinction1)
print(correlation_coefficient)
```

```python
#Code Cell, that makes a scatter plot with r_index_conditioning on the x-axis, and r_index_extinction on the y-axis. Label your axes.

plt.scatter( r_index_conditioning, r_index_extinction)
plt.xlabel('r_index_conditioning')
plt.ylabel('r_index_extinction')
```
