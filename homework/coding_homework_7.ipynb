```python
#Code Cell that imports relevant modules - specifically, numpy as np, and matplotlib.pyplot as plt.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Code Cell that imports these files as numpy arrays called spikes_left_trial1, spikes_right_trial1, spikes_left_trial2, and spikes_right_trial2, respectively.



url1='https://raw.githubusercontent.com/mpompy/Neur265/main/spikes_left_trial1.csv'
spikes_left_trial1=np.loadtxt(url1,delimiter = ',')

url2='https://raw.githubusercontent.com/mpompy/Neur265/main/spikes_right_trial1.csv'
spikes_right_trial1=np.loadtxt(url2,delimiter = ',')

url3='https://raw.githubusercontent.com/mpompy/Neur265/main/spikes_left_trial2.csv'
spikes_left_trial2=np.loadtxt(url3,delimiter = ',')

url4='https://raw.githubusercontent.com/mpompy/Neur265/main/spikes_right_trial2.csv'
spikes_right_trial2=np.loadtxt(url4,delimiter = ',')

```

```python
#Code Cell that prints the length of each array

print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
#Code Cell, write an expression to determine the firing rate of this neuron during each trial (first and second left turn trials, first and second right turn trials).

firing_first_left=len(spikes_left_trial1)/30
firing_second_left=len(spikes_left_trial2)/30
firing_first_right=len(spikes_right_trial1)/30
firing_second_right=len(spikes_right_trial2)/30

print(firing_first_left)
print(firing_second_left)
print(firing_first_right)
print(firing_second_right)
```

```python
#Code Cell, that  makes two subplots, and plot the spike train data from trial 1 in your first subplot, and the spike train data from trial 2 in your second subplot.

fig, axs=plt.subplots(2,1,figsize=(10,10))


axs[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
axs[0].plot(spikes_right_trial1, 1.5 * np.ones_like(spikes_right_trial1), '.')
axs[0].set_xlabel('time (s)')
axs[0].legend({'left', 'right'})
axs[0].set_title('spike train data from trial 1')

axs[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
axs[1].plot(spikes_right_trial2, 1.5 * np.ones_like(spikes_right_trial2), '.')
axs[1].set_xlabel('time (s)')
axs[1].legend({'left', 'right'})
axs[1].set_title('spike train data from trial 2')


```

Does this neuron seem more "bursty", or more regular-firing, during left or right turn trials?

```python
#Code Cell, calculate inter-spike interval (ISI) arrays for each of your four spike arrays using the np.diff() function.

ISIsLeft1 = np.diff(spikes_left_trial1)
ISIsRight1 = np.diff(spikes_right_trial1)
ISIsLeft2 = np.diff(spikes_left_trial2)
ISIsRight2= np.diff(spikes_right_trial2)
```

```python
#Code Cell,that makes four histograms - one for each of your four ISI arrays. Set the bins for your histograms to start at time 0, and end at time 2 s, with a bin size equal to 0.01 s. Label your axes.

fig, axs=plt.subplots(4,1,figsize=(10,10))
plt.subplots_adjust(hspace=1)

bins = np.arange(0,2,0.01)

axs[0].hist(ISIsLeft1, bins, color='red')
axs[0].set_xlabel('ISI [s]')
axs[0].set_ylabel('Counts')
axs[0].set_title('ISIs Left 1')

axs[1].hist(ISIsRight1, bins, color='blue')
axs[1].set_xlabel('ISI [s]')
axs[1].set_ylabel('Counts')
axs[1].set_title('ISIs Right 1')

axs[2].hist(ISIsLeft2, bins, color='pink')
axs[2].set_xlabel('ISI [s]')
axs[2].set_ylabel('Counts')
axs[2].set_title('ISIs Left 2')

axs[3].hist(ISIsRight2, bins, color='green')
axs[3].set_xlabel('ISI [s]')
axs[3].set_ylabel('Counts')
axs[3].set_title('ISIs Right 2')


```

 interpret your histograms. Does this neuron tend to fire in "bursts"? Does it have higher "burstiness" during left or right turn trials?

 * This neuron tends to fire in bursts. It seems to have higher 'burstiness" during the the left turn trials. This is becasue the proportion of small ISIs appears to be  much larger for the left turns.

```python
#Code Cell, calculate the firing rate across time for each of your four arrays.Set your time bins equal to start at 0 s, end at 30 s with a bin size equal to 1 s. Create two subplots, and plot the firing rate over time for each trial in each of the two subplots.Make the line for right-turn trials blue, and the line for left-turn trials red. Label your axes.



binsize = 1
fr_bins = np.arange(0,30,binsize)
fr_left2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2< fr_bins[i+1])
  true_values = spikes_left_trial2[spk_index]
  fr_left2[i] = (len(true_values))/binsize

fr_bins = np.arange(0,30,binsize)
fr_left1 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1< fr_bins[i+1])
  true_values = spikes_left_trial1[spk_index]
  fr_left1[i] = (len(true_values))/binsize

fr_bins = np.arange(0,30,binsize)
fr_right2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2< fr_bins[i+1])
  true_values = spikes_right_trial2[spk_index]
  fr_right2[i] = (len(true_values))/binsize

fr_bins = np.arange(0,30,binsize)
fr_right1 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1< fr_bins[i+1])
  true_values = spikes_right_trial1[spk_index]
  fr_right1[i] = (len(true_values))/binsize


fig, axs=plt.subplots(2,1,figsize=(10,10))
plt.subplots_adjust(hspace=1)

axs[0].plot(fr_bins[:-1], fr_left1,color='red')
axs[0].plot(fr_bins[:-1],fr_right1, color='blue')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Firing Rate [Hz]')
axs[0].set_title('Firing Rate over Time for 1st Trial')


axs[1].plot(fr_bins[:-1], fr_left2,color='red')
axs[1].plot(fr_bins[:-1],fr_right2, color='blue')
axs[1].set_xlabel('Time [s]')
axs[1].set_ylabel('Firing Rate [Hz]')
axs[1].set_title('Firing Rate over Time for 2nd Trial')



```
