```python
#Code Cell that imports relevant modules - specifically, pandas as pd, matplotlib.pyplot as plt, and scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
#Code Cell that imports these files as numpy arrays called acc_hits and acc_fa, respectively

url1='https://raw.githubusercontent.com/mpompy/Neur265/main/acc_hits_hw5.csv'
acc_hits=np.loadtxt(url1,delimiter = ',')

url2= 'https://raw.githubusercontent.com/mpompy/Neur265/main/acc_fa_hw5.csv'
acc_fa=np.loadtxt(url2,delimiter = ',')
```

```python
#Code Cell that prints the shape of each array


print(acc_hits.shape)
print(acc_fa.shape)

```

```python
# Code Cell, that creates variables called time that has the same length as acc_hits and acc_fa, with equally spaced values between "zero" and "four" (the length of each trial, in seconds)

time=np.linspace(0,4,8001)
```

```python
#Code Cell, that creates an expression to determine the sampling rate of the data using the knowledge that you have 8001 samples, and the total length of each sample is 4 seconds

number_of_samples=8001
total_duration=4*8001

sampling_rate=number_of_samples/total_duration

print(sampling_rate)

```

```python
#Code Cell, plot the first trial of acc_hits in one subplot, and the first trial of acc_fa in another subplot. Plot your time variable on the x-axis.


fig, axs=plt.subplots(2,1,figsize=(5,5))

axs[0].plot(time,acc_hits[0,:])
axs[0].set_xlabel('time')
axs[0].set_ylabel('voltage [$\mu V$]')


axs[1].plot(time,acc_fa[0,:])
axs[1].set_xlabel('time')
axs[1].set_ylabel('voltage [$\mu V$]')


plt.show

```

**Describe what you think the dominant oscillation (in Hz) in the LFP might be. Does this differ between the first trial of acc_hits and the first trial of acc_fa?**

- I think the dominant oscillation in the LFP might be right after 2 seconds. This doesn't seem to differ between the the first trial of acc_hits and the first trial of acc_fa.

```python
#Code Cell, calculate and plot the ERP for acc_hits. Put time on the x-axis. Put a vertical black line at the "two" position on the x-axis (the screen touch occurred 2 seconds into the sample on each trial). Put a horizontal black line at the "zero" position on the y-axis. Label your axes

plt.plot(time, acc_hits.mean(0))
plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
plt.title('ERP for acc_hits')

plt.plot([2, 2], [-40,40], 'k', lw=2)

plt.plot([0,4],[0,0], 'k',lw=2)
```

** Interpretation of my ERP for hits. Do you see any evidence of consistent activity in the LFP surrounding screen touch (vertical black bar)**

- I don't see evidence of consitent activity in the LFP surrounding screen touch. The voltage increases and ecreases without any visible pattern. It seems as though a hit causes a drop in voltage in the LFP.

```python
#Code Cell, calculate and plot the ERP for acc_fa. Put time on the x-axis. Put a vertical black line at the "two" position on the x-axis (the screen touch occurred 2 seconds into the sample on each trial). Put a horizontal black line at the "zero" position on the y-axis. Label your axes.

plt.plot(time, acc_fa.mean(0))
plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
plt.title('ERP for acc_fa')

plt.plot([2, 2], [-40,40], 'k', lw=2)

plt.plot([0,4],[0,0], 'k',lw=2)
```
