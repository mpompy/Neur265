```python
#Code Cell that imports relevant modules - specifically, pandas as pd, matplotlib.pyplot as plt, and scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Code Cell that loads expanded file in as a Pandas dataFrame called patch_seq
url= 'https://raw.githubusercontent.com/hallockh/neur_265/main/data/patch_seq_expanded.csv'
patch_seq=pd.read_csv(url)
```

```python
#Code Cell, using the df.plot method from the pandas module, make a bar graph that displays the value counts (how many observations) from the "Marker" column of your patch_seq variable. Label your x and y-axes.

marker_value_counts=patch_seq['Marker'].value_counts()
marker_value_counts.plot(kind='bar')

plt.xlabel('Marker')
plt.ylabel('Counts')

plt.show
```

```python
#Code Cell, repeat step 4, but make a pie chart instead of a bar graph

marker_value_counts=patch_seq['Marker'].value_counts()
marker_value_counts.plot(kind='pie')

plt.xlabel('Marker')
plt.ylabel('Counts')

plt.show
```

```python
#Code Cell, use a for loop to make a scatter plot of the "ISI" (on the x-axis) and "Fast_Trough" (on the y-axis) values in your patch_seq dataframe. Your loop should make data points from "Sst" neurons red, and data points from "Pvalb" neurons blue.

def patchScatter():

  for i in range (len(patch_seq)):

    if patch_seq.iloc[i,2] == 'Sst':
      plt.scatter(patch_seq.iloc[i,0],patch_seq.iloc[i,1],color = "red")

  for i in range(len(patch_seq)):
    if patch_seq.iloc[i,2] == 'Pvalb':
      plt.scatter(patch_seq.iloc[i,0],patch_seq.iloc[i,1],color = "blue")

  plt.xlabel('ISI')
  plt.ylabel('Fast Trough')


plt.show()
```

```python
#Code Cell, create a function called patchScatter() that will output the scatter plot created in step 6 when you call it. Don't forget to tell it to show your scatter plot

patchScatter()
```

```python
#Code Cell, return the mean, standard deviation, and range of the "Fast_Trough" column for "Sst" neurons and "Pvalb" neurons

patch_seq.groupby('Marker')['Fast_Trough'].describe()
```
